```python
import pandas as pd

data = {
    "student_id": ["2026CS001","2026CS002","2026IT001","2026IT002","2026CE001","2026CE002","2026CS003","2026IT003"],
    "name": ["Aditi","Bhavana","Chaitra","Divya","Esha","Fatima","Gauri","Harshada"],
    "department": ["Computer Engineering","Computer Engineering","Information Technology",
                   "Information Technology","Civil Engineering","Civil Engineering",
                   "Computer Engineering","Information Technology"],
    "attendance_pct": [92,64,78,55,88,71,96,43],
    "diary_entries": [12,6,9,4,10,8,14,2],
    "tasks_completed": [8,5,7,3,8,6,9,2],
    "assignment_score": [84,58,72,46,79,66,91,35],
    "lab_score": [88,61,76,52,82,69,94,41]
}

df = pd.DataFrame(data)

print(df.head())
```

This block creates the student dataset using a Python dictionary and converts it into a pandas DataFrame. The first five rows are displayed for verification.

```python
print("Data Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDataset Shape:")
print(df.shape)
```

This block checks the structure of the dataset. It displays data types, counts missing values, and shows the total number of rows and columns.

```python
def readiness_score(attendance, diary, tasks, assignment, lab):

    score = (
        attendance * 0.25 +
        diary * 2 +
        tasks * 3 +
        assignment * 0.20 +
        lab * 0.25
    )

    return round(score, 2)

df["readiness_score"] = df.apply(
    lambda row: readiness_score(
        row["attendance_pct"],
        row["diary_entries"],
        row["tasks_completed"],
        row["assignment_score"],
        row["lab_score"]
    ),
    axis=1
)

print(df[["name", "readiness_score"]])
```

A custom function is created to calculate a readiness score using attendance, diary entries, completed tasks, assignment score, and lab score. The calculated score is stored in a new column.

```python
def classify_student(score):
    if score < 50:
        return "Support"
    elif score < 70:
        return "Developing"
    elif score < 85:
        return "Strong"
    else:
        return "Excellent"

categories = []

for score in df["readiness_score"]:
    categories.append(classify_student(score))

df["category"] = categories

print(df[["name", "readiness_score", "category"]])
```

Students are classified into Support, Developing, Strong, or Excellent categories based on their readiness score using conditional statements and a loop.

```python
print("Department Average Readiness Score:")
print(df.groupby("department")["readiness_score"].mean())

print("\nTop Performer:")
top_student = df.loc[df["readiness_score"].idxmax()]
print(top_student[["name", "department", "readiness_score"]])

print("\nStudents Needing Support:")
support_students = df[df["category"] == "Support"]
print(support_students[["name", "department", "readiness_score"]])
```
